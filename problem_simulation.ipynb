```python
import numpy as np
import random
import math
n = 1
```

```python
n = 10000
s = 0
tries = 1000
ls = []
for i in range(tries):
    centres = []
    radi = []
    for i in range(n):
        c_1, c_2 = random.random(), random.random()
        centres.append([c_1, c_2])
        r = random.uniform(0,np.sqrt(3/ (n * np.pi)))
        radi.append(r)


    x_1, x_2 = random.random(), random.random()
    
    
    for i, e in enumerate(centres):
        if ((x_1 - e[0])**2 + (x_2 - e[1])**2 <= radi[i]**2):
            s += 1 
    s %= 2
    ls.append(s)
    


print(f"Expected value is: {np.mean(ls)}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
mu = 0.1       # Drift
sigma = 0.2    # Volatility
S0 = 100       # Initial stock price
T = 1.0        # Time horizon (in years)
N = 252        # Number of steps (daily)
dt = T / N

# Time array
t = np.linspace(0, T, N)

# Generate Brownian motion
np.random.seed(42)
W = np.random.standard_normal(size=N)
W = np.cumsum(W) * np.sqrt(dt)

# Simulate GBM
S = S0 * np.exp((mu - 0.5 * sigma**2) * t + sigma * W)

# Plot
plt.figure(figsize=(10, 6))
plt.plot(t, S, linewidth=4)
plt.title('Geometric Brownian Motion', fontsize=28)
plt.grid(True)

# Add GBM equation
gbm_eq = r'$S_t = S_0 \exp\left((\mu - \frac{1}{2} \sigma^2)t + \sigma W_t\right)$'
plt.text(0.02, max(S)*0.9, gbm_eq, fontsize=28, bbox=dict(facecolor='white', alpha=0.7))

plt.legend()
plt.tight_layout()
plt.savefig('gbm_example.png', dpi=300)
plt.show()
```
